## –õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

_–ò–º–µ–µ–º_:<br>
$(\sample, \answers)$ - –Ω–∞–±–æ—Ä –æ–±—É—á–∞—é—â–∏—Ö –ø—Ä–∏–º–µ—Ä–æ–≤ (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ N) $\nbObjects$, —Ä–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ $\nbFeatures$
<br>
$|\allY| = \realNumbers$

_–•–æ—Ç–µ–ª–æ—Å—å –±—ã_:<br>
—Å–∞–º—É—é –ø—Ä–æ—Å—Ç–µ–π—à—É—é –º–æ–¥–µ–ª—å, –∫–æ—Ç–æ—Ä–∞—è –±—ã –ø–æ–ø—ã—Ç–∞–ª–∞—Å—å –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä–æ–≤–∞—Ç—å –¥–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è. –ù–æ —á—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–æ—â–µ –ª–∏–Ω–µ–π–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏?
<br><br>
_–ü–æ–∫–æ–Ω–∫—Ä–µ—Ç–Ω–µ–µ_:<br>
–ü—É—Å—Ç—å —É –Ω–∞—Å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è ($\hat{y}$) - –ª–∏–Ω–µ–π–Ω–∞—è –∫–æ–º–±–∏–Ω–∞—Ü–∏—è –≤—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

### –õ–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è

$\hspace{50mm} g(x, \theta) = \hat{y}$, –≥–¥–µ $x$ - –≤—Ö–æ–¥–Ω–æ–π –≤–µ–∫—Ç–æ—Ä, $\hat{y}$ - –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è, $\theta$ - –ø–∞—Ä–∞–º–µ—Ç—Ä—ã. <br>
$\hspace{50mm}$–û–ø—Ç–∏–º–∏–∑–∏—Ä—É–µ–º $L(\hat{y}, y)$.<br>
$\hspace{50mm}$$g(x, \params) = \sum_{i=0}^{M} \params_i * x_{i}$ 

_–í –º–∞—Ç—Ä–∏—á–Ω–æ–º –≤–∏–¥–µ_:<br>
$X * \params = \hat{Y}$<br>

### –ö–∞–∫–∏–µ –º–µ—Ç—Ä–∏–∫–∏ –º–æ–∂–Ω–æ –≤—ã–±—Ä–∞—Ç—å?

MAE (Mean absolute error):
$$MAE = \frac{1}{\nbObjects}\sum_{i=1}^{\nbObjects}\big|\alg(x_i) - y_i\big|$$

RMSE (Root mean squared error):
$$ùëÖùëÄùëÜùê∏ = \sqrt{\frac{1}{\nbObjects}\sum_{i=1}^{\nbObjects}\big(\alg(x_i) - y_i\big)^2}$$

RMSE - —à—Ç—Ä–∞—Ñ—É–µ—Ç –∑–∞ –±–æ–ª—å—à–∏–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è —Å–∏–ª—å–Ω–µ–µ, —á–µ–º –∑–∞ –º–∞–ª–µ–Ω—å–∫–∏–µ( –≤ –æ—Ç–ª–∏—á–∏–∏ –æ—Ç MAE).

MSE (mean squared error):
$$ùëÖùëÄùëÜùê∏ = \frac{1}{\nbObjects}\sum_{i=1}^{\nbObjects}\big(\alg(x_i) - y_i\big)^2$$

–í—ã–±–∏—Ä–∞–µ–º MSE - –ø–æ–ª—É—á–∞–µ–º –º–µ—Ç–æ–¥ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤

_–ò—Ç–æ–≥–æ –ø–æ–ª—É—á–∞–µ–º –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ_:<br>
$(\hat{Y} - Y)^T * (\hat{Y} - Y)\rightarrow min$<br>
$(\sample * \params - \answers)^T * (\sample * \params - \answers)\rightarrow min$

**–ê–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ**
$$(\sample^T * \sample)*\params = \sample^T * \answers\Rightarrow\params = (\sample^T * \sample)^{-1}*\sample^T * \answers$$

$(\sample^T * \sample)^{-1}*\sample^T$ - –ø—Å–µ–≤–¥–æ–æ–±—Ä–∞—Ç–Ω–∞—è –º–∞—Ç—Ä–∏—Ü–∞

–ù–µ –≤—Å–µ–≥–¥–∞ –º–æ–∂–Ω–æ –Ω–∞–π—Ç–∏ –∞–Ω–∞–ª–∏—Ç–∏—á–µ—Å–∫–æ–µ —Ä–µ—à–µ–Ω–∏–µ.<br>
–ü–æ—ç—Ç–æ–º—É –∏—Å–ø–æ–ª—å–∑—É–µ–º **–≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω—ã–π —Å–ø—É—Å–∫**<br>
–ò—Ç–µ—Ä–∞—Ç–∏–≤–Ω—ã–π –º–µ—Ç–æ–¥:<br>
$$\params^{(j+1)}_k:=\params^{(j)}_k - \learningRate * \frac{\partial}{\partial \params^{(j)}_k}\empericRisk$$

–î—è –ú–ù–ö:<br>
$$\empericRisk = \frac{1}{2l} \sum_{i = 1}^l ((\alg(x_i) - y_i)^2$$
$$\frac{\partial}{\partial_k \params}\empericRisk =  \sum_{i = 1}^l ((\alg(x_i) - y_i)* x_{ik}$$
<br>
–ü–æ–≤—Ç–æ—Ä—è—Ç—å, –ø–æ–∫–∞ –Ω–µ —Å–æ–π–¥–µ–º—Å—è:
$$\params^{(j+1)}_k:=\params^{(j)}_k - \learningRate *  \sum_{i = 1}^l ((\alg(x_i) - y_i)* x_{ik}$$
–í –º–∞—Ç—Ä–∏—á–Ω–æ–π —Ñ–æ—Ä–º–µ:
$$\params^{(j+1)}:=\params^{(j)} - \learningRate\frac{1}{m}*(\sample^T*(\sample*\params^{(j)} - Y))$$

### –ú–µ—Ç–æ–¥ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–±–æ–¥–æ–±–∏—è

$p(y_1, y_2, ...y_\nbObjects\:|\:x_1, x_2, ...x_\nbObjects, \params)\rightarrow max_\params$

–ú–µ—Ç–æ–¥ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∞–≤–¥–æ–ø–æ–¥–æ–±–∏—è - –ø—ã—Ç–∞–µ–º—Å—è –ø–æ–ª—É—á–∏—Ç—å —Ç–∞–∫–∏–µ –æ—Ü–µ–Ω–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –¥–µ–ª–∞—é—Ç –Ω–∞—à –Ω–∞–±–æ—Ä –¥–∞–Ω–Ω—ã—Ö –Ω–∞–∏–±–æ–ª–µ–µ –≤–µ—Ä–æ—è—Ç–Ω—ã–º. 

$$W = \prod_i p(y_i\:|\:x_i, \params)\rightarrow max_\params$$

<u>–ü—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ</u>: $y(x_i) = \alg(x_i, \params) + \xi_i$, –≥–¥–µ $\xi_i$ –∏–∑ $N(0, \sigma_i) \rightarrow$
$y_i \sim N(\alg(x_i, \params), \sigma_i)$ 

$$W = \prod_i \frac{1}{\sigma_{i}\sqrt{2\pi}}exp(-\frac{y_{i}-\alg(x_i, \params)}{2\sigma_{i}^2})\rightarrow max_\params$$

–ü–æ–ª—É—á–∞–µ–º:
$$-log(W) = const + \frac{1}{2}\sum_i\frac{1}{\sigma_i^2}(\xi)^2 = const + \frac{1}{2}\sum_i\frac{1}{\sigma^2}\big(\alg(x, \params) - y(x_i)\big)^2\rightarrow min$$

–ü–æ–ª—É—á–∏–ª–∏ –ú–ù–ö (OLS)!

### –†–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è

–ú–ù–ö:
$$MSE \rightarrow min$$

$$MSE + \lambda*\lVert\params\rVert\rightarrow min$$

Ridge –∏–ª–∏ $L_{2}$ - —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è:
$$MSE + \lambda*\lVert\params\rVert^2\rightarrow min$$

–û–¥–Ω–∞–∫–æ, $L_{1}$ –º–æ–∂–µ—Ç –∑–∞–Ω—É–ª–∏—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä—ã, –∞ $L_{2}$ –¥–µ–ª–∞–µ—Ç –∏—Ö –º–∞–ª—ã–º–∏, –Ω–æ –Ω–µ –∑–∞–Ω—É–ª—è–µ—Ç. –ü–æ—á–µ–º—É? 

–ü–æ–¥—Å–∫–∞–∑–∫–∞: –≤–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å –º–µ—Ç–æ–¥–æ–º –≥—Ä–∞–¥–∏–µ–Ω—Ç–Ω–æ–≥–æ —Å–ø—É—Å–∫–∞, –ø–æ—Å—á–∏—Ç–∞–π—Ç–µ–º –ø—Ä–æ–∏–∑–≤–æ–¥–Ω—ã–µ –æ—Ç —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–æ–Ω–Ω—ã—Ö —Ç–µ—Ä–º–∞—Ö –∏ —Å—Ä–∞–≤–Ω–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏—è –ø—Ä–∏ –º–∞–ª—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–∞—Ö $\theta$

# L2 —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è –∫–∞–∫ –ø–æ–º–æ—à–Ω–∏–∫ –≤ –±–æ—Ä—å–±–µ —Å–æ —Å–∫–æ—Ä—Ä–µ–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏

–î–∞–≤–∞–π—Ç–µ –≤—Å–ø–æ–º–Ω–∏–º –ª–∏–Ω–µ–π–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é
$$X*{\params} = Y$$
$$X^TX*{\params} = X^TY$$
$${\params} = (X^TX)^{-1}X^TY$$
–ï—Å–ª–∏ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Å–∫–æ—Ä—Ä–µ–ª–∏—Ä–æ–≤–∞–Ω—ã, —Ç–æ –æ–±—Ä–∞—Ç–Ω—É—é –º–∞—Ç—Ä–∏—Ü—É –±—É–¥–µ—Ç –ø–æ—Å—á–∏—Ç–∞—Ç—å –Ω–µ–ø—Ä–æ—Å—Ç–æ. –ù–∞ –Ω–æ–ª—å –¥–µ–ª–∏—Ç—å –Ω–µ–ª—å—è, –ø–æ–ø—Ä–æ–±—É–µ–º –ø–æ –¥—Ä—É–≥–æ–º—É.

–° –ø—Ä–æ—à–ª—ã—Ö —Å–µ–º–∏–Ω–∞—Ä–æ–≤ –∏ –ª–µ–∫—Ü–∏–π (–ª–∏–±–æ –º–æ–∂–Ω–æ –ø—Ä—è–º —è–≤–Ω–æ –≤—ã–≤–µ—Å—Ç–∏):
$${\nabla}L(X, Y, {\params}) = 0 = X^TX{\params} - X^TY$$

$$L_{reg} = L + 0.5(\lambda\params)^T\lambda\params$$

$${\nabla}L_{reg}(X, Y, {\params}) = {\nabla}((X\params-Y)^T(X\params-Y) + (\lambda\params)^T\lambda\params) = $$
$$ X^TX{\params} - X^TY + 0.5{\nabla}(\lambda\params)^T\lambda\params = (X^TX + \lambda^T\lambda){\params} - X^TY = 0$$

$${\params} = (X^TX+\lambda^T\lambda)^{-1}X^TY$$
