_Пусть_:

$\Population \in \realNumbers^{\nbFeatures}, C = \{-1, +1\}$

$(\sample, \answers)^{\nbObjects}$ - обучающая выборка

_Найти_:

$\params\in\realNumbers^{\nbFeatures}, \params_0\in\realNumbers$

_Тогда_:

$$\alg(x, \params, \params_0) = sign(<x, \params> - \params_0)$$

$<x_1,x_2>$ - скалярное произведение,
$\params $ - нормаль к гиперплоскости, 
$\params_0$ - сдвиг

**Понятие отступа**

$$\margin(x_i) = \margin_i = y_i * (<x_i, \params> - \params_0)$$

$\margin(x_i)$ < 0 тогда, когда ошибаемся на объекте $x_i$

**Почти похоже на геометрического отступ**

$\gamma = \frac{(<x_i, \params> - \params_0)}{||\params||}$

Функционал эмпирического риска:
$$\empericRisk(\params, \params_0, \sample) = \sum_{i = 1}^{\nbObjects}[\margin_i < 0]$$
$Q\rightarrow min$

Сведем к задаче минимизации эмпирического риска:

$$[\margin_i < 0]\leq\loss(M_i)$$


где $\loss:\realNumbers\rightarrow\realNumbers_+$ - непрерывная неубывающая функция.

$$ \sum_{i = 1}^{\nbObjects}\loss(\margin_i)\rightarrow min$$

Выбирая различные виды аппроксимирующих функций потерь мы получаем различные алгоритмы классификации:
* $(1 - \margin)^2$ - Fisher Linear Discriminant
* $(1 - \margin)_+$ - Support Vector Machine
* $(2 * (1 + e^{\margin})^{-1})$ - Perceptron
* $log(1 + e^{-\margin})$ - Logistic Regression

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-20, 20, 100) / 10.0

plt.plot(X, [1 if x < 0 else 0 for x in X], color = [0, 0, 1], label="ER")
plt.plot(X, (1 - X) ** 2, color = [0, 1, 0], label="Fisher Linear Discriminant")
plt.plot(X, [0 if ((1 - x) < 0) else (1 - x) for x in X], color = [1, 0, 0], label="SVM")
plt.plot(X, 2 * (1 + np.exp(X)) ** -1, color = [1, 1, 0], label="perceptron")
plt.plot(X, np.log2(1 + np.exp(-1 * X)), color = [0, 0, 0], label="log regression")

plt.legend()

plt.xlabel("margin")
plt.ylabel("Loss")

plt.show()
```

# Сигмоида (логистическая функция)

Подойдем к вопросу логрегрессии немного по-другому.

Рассмотрим задачу бинарной классификации, то есть задачу пример-ответ $\{X_i, Y_i\}$, где величины $y_i$ могут принимать только значения из множества {0, 1}. Хотим предсказывать, с какой вероятностью ($p_i$) объект $x_i$ принадлежит тому или иному классу.

Вместо вераоятности рассмотрим отношение шансов: 
$$ odds_i = \frac{p_i}{1-p_i}$$
то есть отношение числа благоприятных исходов к неблагоприятным.Но данная величина всегда больше 0, что не подходит нам.

Давайте прологарифмируем, получим величину от $-inf$ до $+inf$
$$logit(p_i) = \eta_i = ln(\frac{p_i}{1-p_i})$$
А полученную функцию будем называть логистическим преобразорванием.
Тогда:
$$ln(\frac{p_i}{1-p_i}) = x_i*{\params}$$ 

$$\eta_i = ln(\frac{p_i}{1-p_i})$$
$$e^{\eta_i}(1-p_i) = p_i$$
$$\frac{e^{\eta_i}}{(1+e^{\eta_i})} = p_i$$
$$p_i = \frac{1.}{(1+e^{-x_i*{\params}})}$$

$$sigmoid(x) = \sigma(x) = \frac{1}{1+exp(-x)}$$

Вероятности при бинароной классификации

$p_{+} = \sigma(x*\theta)$

$p_{-} = 1 - \sigma(x*\theta) = \sigma(-x*\theta)$

$P(y=y_{i}|x_{i}, \theta) = \sigma(M)$

# Метод максимального правдоподобия

$$P\left(\vec{y} \mid X, \vec{w}\right) = \prod_{i=1}^{\ell} P\left(y = y_i \mid \vec{x_i}, \vec{w}\right)$$

\begin{array}{rcl} \log P\left(\vec{y} \mid X, \vec{w}\right) &=& \log \prod_{i=1}^{\ell} P\left(y = y_i \mid \vec{x_i}, \vec{w}\right) \\ &=& \log \prod_{i=1}^{\ell} \sigma(y_i\vec{w}^T\vec{x_i}) \\ &=& \sum_{i=1}^{\ell} \log \sigma(y_i\vec{w}^T\vec{x_i}) \\ &=& \sum_{i=1}^{\ell} \log \frac{1}{1 + \exp^{-y_i\vec{w}^T\vec{x_i}}} \\ &=& - \sum_{i=1}^{\ell} \log (1 + \exp^{-y_i\vec{w}^T\vec{x_i}}) \end{array}

# Бинарная кросс-энтропия

Эмпирический риск:

$$\empericRisk(\params) = -\frac{1}{\nbObjects}\Big[\sum_{i=1}^{\nbObjects}\Big(y^i * log(\alg_{\params}(x_i)) + (1 - y ^i)* log(1 - \alg_{\params}(x^i))\Big)\Big]$$

Используя:

$$\sigma'_x(z(x)) = \sigma(x) * \sigma(1 - x) * z'_x$$
Пусть (упражнение):
$$\frac{\partial}{\partial \params_j}\empericRisk(\params) = \frac{1}{\nbObjects}\sum_{i=1}^{\nbObjects}\big(\alg_{\params}(x^i) - y^i) * x^i_j\big)$$

Градиентный спуск:
$$\params_j := \params_j - \learningRate\frac{1}{\nbObjects}\sum_{i=1}^{\nbObjects}\big(\alg_{\params}(x^i) - y^i) * x^i_j\big)$$

Эти две функции потерь - одно и то же, но просто используют разную разметку даных.
